```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
```

```python
def make_acc(df):
    li = np.array(df['Test acc'].str.split(pat="(").tolist())
    li = np.char.split(li[:,1], ')')
    li = np.array([a[0] for a in li])
    li = li.astype('float')
    return li
```

```python
mixup = pd.read_csv('results/log_ResNet18_cifar10_mixup_4_cross_0.csv', sep='\t')
instahide = pd.read_csv('results/log_ResNet18_cifar10_instahide_4_cross_0.csv', sep='\t')
perturb = pd.read_csv('results/log_ResNet18_cifar10_perturb_4_inside_2.csv', sep='\t')

mixup_ = make_acc(mixup)
instahide_ = make_acc(instahide)
perturb_ = make_acc(perturb)

plt.figure(figsize=(10, 7))

plt.plot(mixup_, label='Mixup', linewidth=3)
plt.plot(instahide_, label='InstaHide', linewidth=3)
plt.plot(perturb_, label='Ours', linewidth=3)
plt.xlabel('EPOCH', fontsize=20)
plt.ylabel('Acc', fontsize=20)
plt.legend(fontsize=20)

plt.show()
```
